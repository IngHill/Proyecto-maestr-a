En la óptica, la elaboración de lentes es un proceso difícil, que implica múltiples pasos como el lapeado de vidrios y el pulido de los mismos, alcanzar el tallado de estas superficies es un método complicado que requiere de la habilidad de un artesano y de un proceso industrial costoso. Los métodos actuales de lapeado y pulido consisten en fricción mecánica de superficies los cuales son de altos costos y de difícil obtención en la región, ya que la finalidad de estos procesos se basa en conseguir que las superficies tengan una rugosidad superficial del orden de micras o incluso de nanómetros; los métodos actuales se ven limitados para alcanzar tal fin.  

En este proyecto se analizará un nuevo método para la creación de superficies ópticas, teniendo un cambio de paradigma en la fabricación de éstas; se considerará la tensión superficial y capilaridad en la formación de meniscos entre dos líquidos examinando la interfaz formada cuando dos líquidos no miscibles entran en contacto. Con el fin de parametrizar las curvas de contorno que forman estas superficies, se hará uso del principio de mínima energía por medio de las ecuaciones Euler-Lagrange. El área de estas superficies se pueden alterar modificando las propiedades de densidad en uno de los fluidos, el cual consistirá en un liquido compuesto o fluido de inmersión, para esto se utilizará una mezcla de  agua con glicerol en diversas concentraciones, esta combinación permite alcanzar una densidad entre 0,997 [g/mL] para agua a temperatura ambiente, hasta 1,263 [g/mL] para glicerol puro. Un segundo líquido o fluido encerrado, será un polímero curable transparente, que puede solidificarse cuando se expone a luz ultravioleta. De esta manera mantendremos la forma que adopte el polímero depositado en un anillo, para un volumen definido, donde las caras de este anillo estarán sujetas a las fuerzas de tensión superficial producidas por el liquido de inmersión. 


Este análisis de las superficies debido a las diferencias de densidades entre los fluidos nos permitirá conocer las curvas de los meniscos, y se espera que presenten las siguientes características: baja rugosidad superficial, invariancia de escala y se puede utilizar para producir incluso lentes de gran tamaño, sin un aumento significativo en el tiempo de fabricación. El alcance de este proyecto será la capacidad de crear fácil y rápidamente componentes ópticos, sin la necesidad de procesos complejos y el uso de equipos costosos, proporcionando nuevas herramientas asequibles para fabricar y probar diseños ópticos.

# 1. Definición mecánica: energía superficial y fuerza capilar.

### 1.1. Trabajo de superficie.
Es bien sabido que el suministro de energía es necesario para crear superficies.
El trabajo para distorsionar un líquido y aumentar su área superficial en una cantidad $dA$.se puede escribir como: $$ W= \gamma d\delta \left(A\right) \text{.}  \;   \;$$

Se define como $\gamma$ es la tensión superficial (o interfacial). Dimensionalmente, $\gamma =\frac{E}{L^2}$.
La tensión superficial $\gamma$ se expresa así en unidades de $ \frac{mJ}{m^2}$. Dicho en palabras,
donde: $\gamma$ es la energía (E) que debe suministrarse por unidad de área (L²).


La tensión superficial también contribuye al trabajo termodinámico. Puede definirse como el aumento de energía interna $U$ o de energía libre $F$ que acompaña a un aumento de superficie: $$\gamma=\left(\frac{\partial \mathbf{F}}{\partial A}\right)_{T,V,n}\text{,}  \;   \;$$
donde n es el numero de moleculas y V es el volumen total del sistema a analizar.
Para más información, en el siguiente link (HTML("[google](https://www.cambridge.org/co/academic/subjects/life-sciences/molecular-biology-biochemistry-and-structural-biology/life-scientists-guide-physical-chemistry?format=PB&isbn=9780521186964)")). 


### 1.2. Fuerzas Capilares.
La tensión superficial también puede verse como una fuerza por unidad de longitud. Dimensionalmente, se puede escribir $\gamma= \frac{F}{L}\text{}  \;   \;$ y se puede expresar $\gamma$ en unidades de $\frac{N}{m}\text{,}  \;   \;$ donde: $$\delta W= \gamma L d(x)\text{.}  \;   \;$$

Donde L es una distancia donde actua la tensión superficial  $\gamma$ en un d(x).

# 2. Superficies curvas esféricas.

Una pompa de jabón, una burbuja de vapor en un líquido en ebullición o las gotas de aceite emulsificado en agua existen gracias a la diferencia de presiones interna y externa y a la tensión superficial del líquido.Vease Figura 1.


$$Figura \; 1.$$

En la expansión o contracción de la superficie esférica en una fase fluida se encuentran involucrados dos trabajos: el de la fase volumétrica y el de la superficie. 
La ecuación general de Young-Laplace para un sistema de dos fases y un solo componente puede construirse a partir de la energía libre de Helmholtz que contiene los trabajos de superficie $(dA)$ y de volumen $(PdV)$

$$dF=-SdT-P^\alpha dV^\alpha -P^\beta dV^\beta + \gamma  dA=0 \text{,}  \;   \;(1)$$ 

Donde $dF$ es la variación de energía libre de Helmholtz, $S$ la entropía, $T$ la temperatura absoluta, $P^\alpha$ y $P^\beta$ las presiones en cada fase (burbuja interna y exterior de la burbuja), $dV^\alpha$ y $dV^\beta$ los cambios de volumen, $\gamma$ la tensión superficial en la superficie de la burbuja y $dA$ la variación de área superficial.

En el equilibrio $dF = 0$, y bajo condiciones isotérmicas $dT=0$; la ecuación de la energia libre queda de la forma:

$$P^\alpha dV^\alpha + P^\beta dV^\beta = \gamma dA \text{.}  \;   \;(2)$$

El incremento de volumen de una fase se efectúa a expensas del decremento del volumen de la otra, de modo que:
$$-dV^\alpha=dV^\beta\text{.}  \;   \;$$

Reemplazando en la ecuación anterior (2) tenemos: $$(P^\alpha-P^\beta)dV=\Delta PdV= \gamma dA\text{,}  \;   \; $$

$$\Delta P=\gamma \frac{dA}{dV}\text{.}  \;   \;(3)$$

El incremento de volumen producido por la expansión ocurre simultáneamente con el aumento de área, de modo que en equilibrio y para el caso de una esfera de radio $(R)$ donde los cambios de volumen y de área están dados por $dV=4πR2$
$dR$ y $dA = 8πRdR$, la diferencia de presiones es: $$\Delta P= \frac{2\gamma}{R}\text{,}  \;   \;(4)$$

La expresión anterior corresponde a la ecuación general de Young-Laplace (4) para la condición esférica, e indica que la diferencia de presión a través de una superficie depende de la tensión superficial de la película y del radio de la esfera. 

# 3. Superficies curvas no esféricas.
La ecuación general de Young-Laplace puede extenderse a superficies curvas no esféricas,
considerando que éstas requieren dos radios máximos de curvatura para su correcta descripción.vease Figura 2.


$$Figura\;  2.$$

De acuerdo con la ecuación la diferencia de presión en cada plano es:$$\Delta P_1= \frac{2\gamma}{R_1}\text{,}  \;   \;$$   $$\Delta P_2= \frac{2\gamma}{R_2}\text{,}  \;   \;$$ 

la diferencia de presión es igual en toda la superficie de modo que:$$\Delta P=(P^\alpha-P^\beta)= (\Delta P_1 + \Delta P_2)=2 \Delta P\text{,}  \;   \;$$

$$2\Delta P=\frac{2\gamma}{R_1}+\frac{2\gamma}{R_2}\text{,}  \;   \;(5)$$

obteniendo el delta de presión y sustituyendo en la ecuación particular para una esfera obtenemos la ecuación general de Young-Laplace para superficies curvas no esféricas:

$$\Delta P=\gamma\left(\frac{1}{R_1}+\frac{1}{R_2}\right)=\gamma C\text{,}  \;   \;(6)$$ 

donde $C=Curvatura\text{.}$

# 4. Ecuación de Young-Laplace a otras formas geométricas.

### 4.1. Curvatura y radio de curvatura.
La curvatura (C) de una sección de arco en un punto arbitrario (A), se define como el límite de
la razón del ángulo (ø) de la tangente en (A), respecto a la longitud del arco (S);vease Figura 3 o en el siguiente enlace (HTML("[google](https://intranetua.uantof.cl/estudiomat/calculo3/stewart.pdf)")).

$$ Figura\; 3.$$

Definimos que uno sobre el radio o la curvatura, se define como  el desplazamiento del angulo sobre la curva :$$1/(\rho)=\frac{\triangle\theta}{\triangle(s)}=    Curvatura\text{.}  \;   \;(7)$$


Hacercando al limite esta relación nos da: $$1/\rho= \lim_{\triangle(s)\rightarrow 0}(\frac{\triangle\theta}{\triangle(s)})=\frac{d\theta}{d(s)}\text{,}  \;   \;$$

la curvatura la podemos escribir de otra forma utilizando la regla de la cadena: $$\frac{d\theta}{d(s)}=\frac{d\theta}{d(x)}*\frac{d(x)}{d(s)}\text{,}  \;   \;(8)$$

a comodando todo en función de $d(x)$: $$\frac{d\theta}{d(s)}=\frac{d\theta}{d(x)}*\frac{1}{\frac{d(s)}{d(x)}}\text{.}  \;   \;  \;   \;(8.1)$$

Analizamos que el ángulo $\theta$ se puede conocer como la tangente del angulo:  $$\tan (\theta) = f'(x_0)\,; \,\,\mbox{la derivada de la función}\,, $$ es decir,

Si $θ(x)$ es el ángulo, medido en radianes, que
la tangente a la curva elástica forma con la horizontal en A (figura 3), y al recordar
que este ángulo es pequeño: $$\theta\cong \tan (\theta) = \frac{d(y)}{d(x)}\text{,}  \;   \;(9)$$

$$\theta = \tan{\frac{d(y)}{d(x)}}\text{.}  \;   \;$$

Por lo tanto podemos escribir a $\frac{d\theta}{d(x)}$ como:$$\frac{d\theta}{d(x)}=\frac{d}{d(x)}*\tan{y'}=\frac{y''}{1+(y')^2}\text{.}  \;   \;(10)$$

Para encontrar el termino $\frac{d(s)}{d(x)}$ se obtiene de la definición de longitud de arco, la cual esta dada como: $$d(s)=\sqrt{1+f´(x)^2}dx\text{,}  \;  \;(10.1)$$$$

$$\frac{d(s)}{d(x)}=(1+(y')^2)^\frac{1}{2}\text{,}  \;   \;(11)$$

al retomar la ecuación $\frac{d\theta}{d(s)}=\frac{d\theta}{d(x)}*\frac{1}{\frac{d(s)}{d(x)}}$ nos queda: $$C=\frac{1}{\rho}=\frac{d\theta}{d(s)}=\frac{y''}{(1+(y')^2)^\frac{3}{2}}\text{.}  \;   \;(12)$$

De esta manera hemos encontrado una función para la descrivir la curvatura  

### 4.2. Ecuación de Young-Laplace bajo campos externos.
En ausencia de campos externos, las superficies adquieren la forma esférica, mientras que bajo la influencia de un campo, como por ejemplo el gravitacional, una superficie esférica sufre deformación.
Tomemos el caso de una gota pendiente de una boquilla como la que se representa en la figura 4.


$$Figura\; 4.$$ (HTML("[google](https://link.springer.com/book/10.1007/978-0-387-21656-0)")) 

Al abrir un grifo, es común observar que el chorro de agua se rompe en gotitas para disminuir su energía superficial. Para comprender mejor los beneficios de un área de superficie más baja, el radio de curvatura es el de una esfera de radio $( Ro)$; a partir de ese punto la curvatura y los radios de curvatura cambian punto a punto con la presión hidrostática.
La diferencia de presiones es: $$\Delta P =\frac{2\gamma}{R_0}+\Delta \rho g y \text{,}  \;   \;(13)$$

donde: $Δρ$ es la diferencia de densidades entre la fase líquida y gaseosa y $\mathrm{g}$ la aceleración de la gravedad. $$\;$$
El primer término a la derecha en la ecuación se refiere a la contribución por curvatura (esférica) a la diferencia de presión, mientras que el segundo término representa la contribución y deformación de la condición esférica por la presión hidrostática provocada por el campo externo, en este caso, gravitacional; así la condición no esférica se construye a partir de la condición esférica sometida al campo externo, combinando este término con las ecuaciones Young-Laplace: $$\gamma \left(\frac{1}{R_1} + \frac{1}{R_2}\right) =\frac{2\gamma}{R_0}+\Delta \rho g y\text{.}  \;   \;(14)$$

Sí dividimos entre la tensión superficial $\gamma$  y multiplicamos por el radio $R_0$ la ecuación (14), obtenemos:

$$ \;    \;  $$
\begin{eqnarray}
&\left(\frac{1}{R^*_1} + \frac{1}{R^*_2}\right) =2+\frac{\Delta \rho g R_0²}{\gamma}Y\text{;} \; (14)\\
& R^*_i=\frac{R_0}{R_i} \, ,\; i=1,2\qquad \& \qquad Y=\frac{y}{R_o}
\end{eqnarray}

En donde el ultimo termino de $\frac{\Delta \rho g R_0²}{\gamma}$ es el número de bond reúne las propiedades del fluido y se incluye en el parámetro de forma y $Y$ es la altura normalizada.

De las ecuaciones de curvatura podemos obtener de la ecuación 8.1: $$\frac{d\theta}{ds}=\frac{d\theta}{dx} \frac{1}{\frac{ds}{dx}}\text{.}  \;   \;$$

El $\frac{d\theta}{dx}$ se puede calcular teniendo en cuenta las siguientes relaciones $dz=ds\sin{\theta}$ y $dx=ds\cos{\theta}$,

# 6 Otras formas (Método variacional)

Para conseguir la minima energía en superficies, se puede obtener por medio del metodo de variaciones el cual consiste en obtener el funcional del sistema, para realizar este metodo, retomaremos el problema de la gota suspendida en la fibra.



para construir nuesro funcional lo obtendremos del balance de energía, obteniendo la energía libre de gibbs $G$: $$ G= H -TS\text{.} \;(28)$$

Donde la energía libre de Gibbs es igual a la entropia menos la temperatura por la entropia.  

Realizamos un balance de energia interna al fluido obtenemos:

$$ \Delta {U} = dq+dw  \text{.}\;(29)$$

Donde la energía interna es igual a la variación de calor más trabajo, la variación del calor en un proceso reversible en función de la entropía da:


$$dq=Tds \text{.} \;(30)$$

Como el trabajo es solo la presión por el volumen y debido al área superficial se tiene:

$$dw= -p(dv)+ \gamma dA \text{.} \;(31)$$



Sustituyendo las ecuaciones 30 y 31, en la ecuación 29 se obtiene la variación de la energía interna:

$$ dU=T(ds)-P(dv)+\gamma dA \text{.}$$


Definimos la energía libre de Gibbs de la ecuación 28 como: 

$$G= H -TS = (U+PV)-TS \text{.} $$

Derivando la ecuación de energía libre:

$$dG=dU+Pdv+vdP-Tds-sdT \text{.}$$

Reemplazando por la energía interna, la derivadad de Gibbs nos queda:

$$dG=VdP-sdT+\gamma dA \text{.}$$

Al asumir que el proceso ocurre a temperatura constante (isotérmico), la ecuación 28 queda definida como:

$$ dG=Vdp+\gamma dA \text{.}$$

Las superficies mínimas se pueden calcular con la ayuda de la fórmula de Laplace.
Alternativamente, se puede minimizar la superficie, manteniendo el volumen constante, por medio de las ecuaciones de Euler-Lagrange.

### Ecuación Euler-Lagrangre (meétodo variacional)
La importancia de las interfases curvas se extiende a áreas como la estabilidad de espumas (Rosen 1989), donde la redistribución de tamaño de las burbujas (que es una consecuencia de la presión interna de la burbuja) depende de la curvatura de sus paredes (Bikerman, 1965), son también importantes en el efecto de la curvatura sobre la energía libre molar de una sustancia y su presión de vapor (Adamson, 1997, y Hammel, 1958), así la ecuación que muestra este
efecto se construye a partir de la influencia de la variación de la presión sobre la energía libre.

miraremos como una función puede minimizarce para saber entender la superficie minima en nuestro problema.
Dada la función $f(x,y,z)$  la minimización de esa función estará dada por $$F(y)=\int_{x_0}^{x_1} f\left(x,y(x),y'(x)\right)dx \text{.} $$

Entre las funciones que tienen los estremos fijos, con $y(x_0)=y_0 \text{;} y(x_1)=y_1 \text{,}$

si se escoje una función h(x) cuales quiera, que vale cero en los extremo de manera qe:
$$h(x_0)=h(x_1)=0 \text{.}$$

Entonces la función $$F(y+\varepsilon h) \text{.}$$  

Tendra los mismos valores en los extremos que mi función $f(y)$ y la función $F(y+\varepsilon h)$ tiende al minimo cuando $\varepsilon$ tiende a cero.

 $$\frac{\partial }{\partial \varepsilon} \mid_\varepsilon{=_0} F(y+\varepsilon h)=0 \text{.} $$

Al integrar la función:

$$\int_{x_0}^{x_1}\left(\frac{\partial}{\partial \varepsilon} f(x,y(x)+\varepsilon h(x),y'(x)+\varepsilon h'(x))\right) dx=\int_{x_0}^{x_1}\left(\frac{\partial f}{\partial y}h(x)+\frac{\partial f}{\partial z}h'(x)\right)dx=G \text{.}$$

Nos centramos en el segundo termino de la integral, al evaluar por partes nos que da: $$\int_{x_0}^{x_1}\left(\frac{\partial f}{\partial z}h'(x)\right)dx=\frac{\partial f}{\partial z}h'(x) \Big |_{x_0}^{x_1} - \int_{x_0}^{x_1}\left(\frac{\partial }{\partial x} \frac{\partial f }{\partial z}\right)h(x)dx \text{.} $$

Como $h(x)$ no varia en los extremos la primera parte de la integral da 0, la función G que da: $$G=\int_{x_0}^{x_1}\left(\frac{\partial f}{\partial y} -\frac{\partial }{\partial x} \frac{\partial f }{\partial z}\right)h(x)dx=0 \text{.} $$

Por lo tanto: $$\left(\frac{\partial f}{\partial y} -\frac{\partial }{\partial x} \frac{\partial f }{\partial z}\right)=0 \text{.}$$

Esta expresión es conocida como la Ecuación Euler-Lagrange, para el caso particular de la curva en la gota la ecuación no depende de x y se puede utilizar la expresión de Beltromi: $$f-y'\frac{\partial f}{\partial z}=constante \text {.}$$

La expresión para nuestro ejercicio estará dada como: $$\frac{\partial f }{\partial z}-\frac{d}{dx}\frac{\partial f}{\partial \dot z}=0 \text{.}$$

Resolviendo en  Sympy: $$ f(z,\dot{z})= ( 2\pi \gamma z\sqrt{1+(\dot z)²})- (\Delta{p} \pi (z²-b²)) \text{.} $$
