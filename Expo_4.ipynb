Aplicamos el teorema de linealización:

$$ \mathbb{J} = \begin{bmatrix}
\frac{\partial f}{\partial x} & \frac{\partial f}{\partial y} \\
\frac{\partial g}{\partial x} & \frac{\partial g}{\partial y}
\end{bmatrix} $$

Donde $f$ y $g$ son las funciones de las ecuaciones diferenciales (en este caso, $f = 14x-\frac{x²}{2} -xy$ y $g = 16y -\frac{y²}{2} -xy$).

Sustituyendo las derivadas en la matriz Jacobiana, tenemos:

$$ \mathbb{J} = \begin{bmatrix}
14-x-y & -x \\
-y & 16-y-x
\end{bmatrix} $$ 

Luego 
$$ \begin{bmatrix} \dot{x} \\ \dot{y} \end{bmatrix}\approx \mathbb{J} \begin{bmatrix} x \\ y \end{bmatrix}$$ 

Para el análisis de estabilidad del modelo se usa el teorema de linealización de Liapunov y Poincaréq.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las ecuaciones del sistema
def sistema(y, t):
    x, y = y
    dxdt = 14*x - (1/2)*x**2 - x*y
    dydt = 16*y - (1/2)*y**2 - x*y
    return [dxdt, dydt]

# Crear una malla de puntos en el plano xy
x = np.linspace(-50, 50, 50)
y = np.linspace(-50, 50, 50)
X, Y = np.meshgrid(x, y)

# Calcular las derivadas en cada punto de la malla
DX, DY = sistema([X, Y], 0)

# Graficar el campo vectorial y las isoclinas
fig, ax = plt.subplots(figsize=(8, 8))
ax.quiver(X, Y, DX, DY, color='r', alpha=0.5)
ax.contour(X, Y, DX, levels=[0], colors='b', linestyles='dashed')
ax.contour(X, Y, DY, levels=[0], colors='g', linestyles='dashed')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim([-50, 50])
ax.set_ylim([-50, 50])
plt.grid()

# Resolver el sistema de ecuaciones con condiciones iniciales
from scipy.integrate import odeint
y0 = [1, 1]
t = np.linspace(0, 1, 10)
sol = odeint(sistema, y0, t)
ax.plot(sol[:,0], sol[:,1], 'b')
ax.plot(y0[0], y0[1], 'go')

plt.show()
```

## 2) Estudiar los puntos crı́ticos y la solución de la ED autónoma:



$$ m* \ddot{x} = - k*x $$



Donde 
$$w_0² = \frac{k}{m}$$

Frecuencia temporal 

reemplazando tenemos 

$$ \ddot{x} + w_0x = 0 $$

Linealizando la ED 

$$ \ddot{x} = \frac{d\dot{x}}{t} = \frac{d\dot{x}}{dx}.\frac{dx}{t}= \frac{d\dot{x}}{x}.\dot{x}$$



reemplazando 

$$ \frac{d\frac{\dot{x}²}{2}}{dx} = -w_0²x$$


Integrando a ambos lados de la ecuación:

$$ \frac{dx}{dt} = \dot{x} = \pm \sqrt{w_0²x²+c}$$

E.D. de primer orden de primer grado, autonoma.

Análisis de puntos críticos:



$$ \frac{dx}{dt} = 0 $$

reemplazando:


$$ x =  \pm \sqrt{\frac{c}{w_0}} $$

La solución de la ecuación diferencial por medio de Laplace es:

$$ m\ddot{x} + k*x = 0$$

$$\mathcal{L}\lbrace m \ddot{x} \rbrace + \mathcal{L}\lbrace kx \rbrace = 0 $$

Donde 

$$ \mathcal{L}\lbrace \ddot{x} \rbrace =  s²\mathcal{L}\lbrace x \rbrace -sx(0)- \dot{x}(0) $$

Al reemplazar 

$$ ms²\mathcal{L}\lbrace x \rbrace - msx_0 -mv_0 + k\mathcal{L}\lbrace x \rbrace = 0  $$

$$ (ms² + k) \mathcal{L}\lbrace x \rbrace = msx_0 + mv_0 $$


$$ x = \mathcal{L}⁻1\lbrace \frac{msx_0}{ms²+k} + \frac{mv_0}{ms²+k} \rbrace  $$

$$ x = x_0 \mathcal{L}⁻1\lbrace \frac{s}{s²+\frac{k}{m}}\rbrace + v_0 \mathcal{L}⁻1\lbrace \frac{1}{s²+\frac{k}{m}} \rbrace  $$

utilizando el inverso de Laplace, la solución encontrada es: 

$$ x = x_0 cos(w_0t) + \frac{v_0}{w_0}sen (w_0t) $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la ecuación
C = 1.0
K = 0.9
m = 1.0
  
# Condiciones iniciales
x0 = 0.0
t0 = 0.0

# Paso temporal y número de puntos
dt = 0.01
num_steps = 1000

# Definimos la función f(t,x) = dx/dt
def f(t, x):
    return np.sqrt((K/m) * x**2 + C)

# Creamos arrays para almacenar la solución
t = np.zeros(num_steps)
x = np.zeros(num_steps)

# Asignamos las condiciones iniciales
t[0] = t0
x[0] = x0

# Aplicamos el método de Euler para obtener la solución numérica
for i in range(num_steps - 1):
    t[i+1] = t[i] + dt
    x[i+1] = x[i] + dt * f(t[i], x[i])

# Graficamos la solución
plt.plot(t, x)
plt.xlabel('Tiempo')
plt.ylabel('Distancia')
plt.grid(True)
plt.show()
```

## 3) Describir y/o analizar la bifurcación de las EDs

Describir y/o analizar la bifurcación de las EDs

$$ \frac{dx}{dt} = \mu x + x3 $$

$$ \frac{dy}{dt} = -y $$


Ayuda: Revisar bifurcación Pitchfork.

Para encontrar los puntos críticos de este sistema de ecuaciones diferenciales, igualamos las derivadas a cero y resolvemos las ecuaciones simultáneamente.

Para la primera ecuación:
$$    \frac{dx}{dt} = \mu x + x³= 0 $$

Esto implica que uno de los factores debe ser cero:

$   x=0 $ o  $ \mu x + x³ = 0    $


no importa el valor de $y$, la derivada de $y$ será siempre cero, lo que significa que no es un punto crítico interesante.

Por lo tanto, nos enfocaremos en la segunda posibilidad:

$$ \mu + x² = 0 ⇒ x² = -\mu ⇒  x = \pm \sqrt{\mu} $$


Para la segunda ecuación:    

$$ \frac{dy}{dt} = - y $$

Esto implica que $y = 0$.

Por lo tanto, los puntos críticos del sistema de ecuaciones son $(x, y) = P0(\sqrt{\mu}, 0)$ y $P1(-\sqrt{\mu}, 0)$

Análisis de $ \mu $:



1) Para $\mu < 0$:

En este caso, tenemos la expresión $-\sqrt{\mu}$, lo que significa que $\sqrt{\mu}$ es un número imaginario. Dado que los puntos críticos deben tener valores reales, no hay puntos críticos en este caso.



2) Cuando $\mu = 0$:

Si $\mu = 0$, la ecuación se convierte en: $x=0$
En este caso, tenemos un único punto crítico en $x = 0$. Sin embargo, hay que tener en cuenta que la segunda ecuación sigue siendo $y=0$, lo que significa que hay infinitos puntos críticos en el eje $x$.



3) Cuando $\mu > 0$:

Para $\mu > 0$, la expresión $\sqrt{\mu}$ es un número real. Por lo tanto, tenemos dos puntos críticos simétricos:

$(x,y)=(\sqrt{\mu},0)$ y $(x,y)=(-\sqrt{\mu},0)$



se concluye:

  Cuando $\mu < 0$, no hay puntos críticos reales.
  Cuando $\mu = 0$, hay un único punto crítico en $x = 0$.    Cuando $\mu > 0$, hay dos puntos críticos simétricos en $(\sqrt{\mu}, 0)$ y $(-\sqrt{\mu}, 0)$.



```python
import numpy as np
import matplotlib.pyplot as plt

def dx_dt(x, y, mu):
    return mu * x + x**3

def dy_dt(x, y):
    return -y

# Parámetros
mu_values = np.linspace(-2, 2, 1000)  # Valores de mu para la bifurcación
x_values = np.linspace(-2, 2, 100)    # Valores de x para el espacio de fase
y_values = np.linspace(-2, 2, 100)    # Valores de y para el espacio de fase

# Bifurcación
plt.figure(figsize=(8, 6))
for mu in mu_values:
    x_critical = np.sqrt(-mu)
    plt.plot(mu, x_critical, 'bo', markersize=2)
    plt.plot(mu, -x_critical, 'bo', markersize=2)

plt.xlabel('mu')
plt.ylabel('x')
plt.title('Bifurcación de Pitchfork')
plt.grid(True)
plt.show()

```

```python
# Espacio de fase
plt.figure(figsize=(8, 6))
for x in x_values:
    for y in y_values:
        plt.quiver(x, y, dx_dt(x, y, 1), dy_dt(x, y), color='b', alpha=0.5)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Espacio de fase')
plt.grid(True)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def dx_dt(x, y, mu):
    return mu * x + x**3

def dy_dt(x, y):
    return -y

def find_critical_points(mu):
    x_critical = np.sqrt(-mu)
    return [(x_critical, 0), (-x_critical, 0)]

# Parámetros
mu = 1.0  # Valor de mu (puedes ajustarlo según tus necesidades)
x_range = np.linspace(-2, 2, 100)  # Valores de x para el espacio de fase
y_range = np.linspace(-2, 2, 100)  # Valores de y para el espacio de fase

# Encontrar puntos críticos
critical_points = find_critical_points(mu)

# Crear malla de valores para el espacio de fase
X, Y = np.meshgrid(x_range, y_range)
U = dx_dt(X, Y, mu)
V = dy_dt(X, Y)

# Graficar el espacio de fase tridimensional
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, np.zeros_like(X), U, V, np.zeros_like(X), color='b', alpha=0.5)

# Puntos críticos
for point in critical_points:
    ax.plot([point[0]], [point[1]], [0], 'ro', markersize=5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('dz/dt')
ax.set_title('Espacio de fase tridimensional (mu > 0)')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def dx_dt(x, y, mu):
    return mu * x + x**3

def dy_dt(x, y):
    return -y

# Parámetros
mu_range = np.linspace(-2, 2, 1000)  # Valores de mu para la bifurcación
x_range = np.linspace(-2, 2, 100)    # Valores de x para el espacio de fase
y_range = np.linspace(-2, 2, 100)    # Valores de y para el espacio de fase

# Crear mallas de valores para la bifurcación
MU, X = np.meshgrid(mu_range, x_range)
Y = np.zeros_like(MU)
for i, mu in enumerate(mu_range):
    Y[:, i] = np.sqrt(-mu)

# Graficar la bifurcación en 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(MU, X, Y, color='b', alpha=0.5)

ax.set_xlabel('mu')
ax.set_ylabel('x')
ax.set_zlabel('y')
ax.set_title('Bifurcación de Pitchfork en 3D')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)
plt.show()
```
