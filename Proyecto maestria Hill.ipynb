# Minimizaci칩n de la energ칤a libre

Miraremos como una funci칩n puede minimizarce para entender la superficie minima en nuestro an치lisis. Dada la funci칩n de Gibbs el funcional de pontenciales termodimadin치micos bajo la integral estar치 dada por la funci칩n $f(r,h, h_r,h_{\theta})$  de la siguiente manera:  

$$\begin{equation}
\tag{7} J=\int_{r_0}^{r_1} f\left(r,h,h_{r}, h_{\theta}; r, \theta\right)drd\theta \text{.}
\end{equation}$$

Utilizaremos el metodo de calculo variacional  $\color{green}{\text{[6]}}$ para encontrar la funci칩n que nos minimize definiendo $h$ como: 

$$h(r, \theta, 톝) = h(r,\theta, 0) + 톝H(r,\theta),$$ d칩nde $H(r,\theta)$ es cero cuando se encuentra en los extremos $r_1$ y $r_2$, para que resulte un extremal ha de cumplirse que $\frac{객J}{객톝} = 0$.

Que al aplicar esta condici칩n a la ecuaci칩n 7 se tiene que:

$$ \frac{\partial J}{\partial 톝} = \int_{r_0}^{r_1}\left(\frac{\partial}{\partial \varepsilon} f(r,h,h_{r}, h_{\theta}; r, \theta)\right)dr = \int_{r_0}^{r_1}\left(\frac{\partial f}{\partial h}H(r, 풪) + \frac{\partial f}{\partial h}\frac{\partial H}{\partial r} + \frac{\partial f}{\partial h_{\theta}}\frac{\partial H}{\partial \theta}\right)dr
\text{.}$$

Nos centramos en el segundo termino de la integral, al evaluar por partes nos que da: $$\int_{r_0}^{r_1}\left(\frac{\partial f}{\partial h}\frac{\partial H}{\partial r}\right)dr = \frac{\partial f}{\partial h}H(r, \theta) \Big |_{r_0}^{r_1} - \int_{r_0}^{r_1}\frac{\partial }{\partial r} \left(\frac{\partial f }{\partial h_r} \right)H(r, \theta)dr \text{.} $$

Como $H(r, \theta)$ se anula en los extremos la primera parte de la integral es 0, siguiendo el mismo razonamiento para el tercer termino de la integral original, se llega a: $$\frac{\partial J}{\partial 핏}=\int_{r_0}^{r_1}\left(\frac{\partial f}{\partial h} -\frac{\partial }{\partial r} \frac{\partial f }{\partial h_r} - \frac{\partial}{\partial \theta}\frac{\partial f}{\partial h_{\theta}}\right)H(r, \theta)dr=0 \text{.} $$

Por lo tanto: 
$$\begin{equation}
\frac{\partial f}{\partial h} -\frac{\partial }{\partial r} \frac{\partial f }{\partial h_r} - \frac{\partial}{\partial \theta}\frac{\partial f}{\partial h_{\theta}}=0 \text{.}
\end{equation}$$


Al existir simetria de rotaci칩n podemos omitir el ultimo t칠rmino dando como resultado la ecuaci칩n de Euler-lagrange:


$$\begin{equation}
\tag{8.0} \frac{\partial f}{\partial h} -\frac{\partial }{\partial r} \frac{\partial f }{\partial h_r} =0 \text{.}
\end{equation}$$

La soluci칩n por medio de la ecuaci칩n de Euler-lagrange se lograra realizandose dos veses para las dos superficies dando como resultrado un sistema de ecuaciones diferenciales acopladas en terminos de $h^t$ y $h^b$ donde f es: 

$$\begin{equation}
f= \left( \sqrt{1+(h_{r}^t)^2} + \sqrt{1+(h_{r}^b)^2}- \frac{\Delta \rho g}{\gamma} \left((h^t)- (h^b)쑆ight) +\lambda_{1}\frac{(h^t-h^b)}{\gamma}+\lambda_{2}\frac{(h^b)}{\gamma}\right) \, r\;dr \text{.}
\end{equation}$$

```python
def f(h_t, h_b, h_rt,h_rb, h_0t, h_0b, r,
      delta_P = Symbol("\\Delta P"), g = Symbol("g"), gamma = Symbol("\\gamma"),
      lambda1 = Symbol("\\lambda_1"), lambda2 = Symbol("\\lambda_2")):
  
  return (sqrt(1 + h_rt**2 + h_0t**2/r**2) + sqrt(1 + h_rb**2 + h_0t**2/r**2) - delta_P*g/(2*gamma)*(h_t**2 - h_b**2) + lambda1*(h_t - h_b)/gamma + lambda2*h_b/gamma)*r

r, theta = symbols("r \\theta")

#Caso h(r, theta).--------
h_t = Function("h^t")(r, theta); h_rt = diff(h_t, r);  h_0t = diff(h_t, theta)
h_b = Function("h^b")(r,theta); h_rb = diff(h_t, theta);  h_0b = diff(h_b, theta)
funcional = f(h_t, h_b, h_rt,h_rb, h_0t, h_0b, r)

#Caso independiente de theta h(r).
h_t = Function("h^t")(r); h_rt = diff(h_t, r);
h_b = Function("h^b")(r); h_rb = diff(h_b, r)
funcional = f(h_t, h_b, h_rt,h_rb, 0, 0, r) #Funcional.

#Ecuaciones diferenciales apartir de Euler-Lagrange.
PDE1 = Eq(diff(funcional, h_t) - simplify(diff(diff(funcional, h_rt), r)) - simplify(diff(diff(funcional, h_0t), theta), 0)) #PDE para h_t
PDE2 = Eq(diff(funcional, h_b) - simplify(diff(diff(funcional, h_rb), r)) - simplify(diff(diff(funcional, h_0b), theta), 0)) #PDE para h_b
```
