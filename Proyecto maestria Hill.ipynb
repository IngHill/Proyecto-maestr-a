# Estudio de la generación de superficies en fluidos inmersos en soluciones no miscibles

##  CAPILARIDAD 

La capilaridad se refiere a las interfaces que son lo suficientemente móviles para adoptar formas en equilibrio, ejemplos comunes de esto son los meniscos, las gotas de agua cuando llueve, una gota de aceite en agua. Dado que estas formas estan en equilibrio la capilaridad ocupa un lugar en el estudio termodinámico en relación macroscopica  y no en el entorno molecular. Por lo tanto definiremos las fuerzas implicadas para tal fin. 

## Tensión superficial 

El suministro de energía necesario para crear superficies, es el trabajo para distorsionar un líquido y aumentar su área superficial en una cantidad $[𝑑𝐴]$,  $\color{green}{\text{[1]}}$ se puede escribir como:

$$ \delta W = \gamma  dA\ \text{.}  \;   \;$$

Donde $\gamma$ se define como la tensión superficial o interfacial y esta definida como energía que debe suministrarse por área y esta dada en unidades S.I de $ \frac{J}{m^2}$. La tensión superficial también contribuye al trabajo termodinámico y puede definirse como el aumento de energía interna $U$ o de energía libre $G$ el cual acompaña a un aumento de superficie:

$$\gamma=\left(\frac{\partial \mathbf{F}}{\partial A}\right)_{T,V,n}\text{,}  \;   \;$$

La tensión superficial depende de propiedades como la temperatura $T$ el volumen $V$ y el numero de moles $n$ cuando se tienen potenciales quimicos  $\color{green}{\text{[2]}}$.

# Energía libre
 
A continuación estudiaremos un proceso isotérmico real (irreversible) en el que tanto el sistema como los alrededores se mantienen a la temperatura T. Al mismo tiempo definiremos este proceso como un proceso a volumen constante que no depende del tiempo. El proceso que estudiaremos es el proceso de mezcla de dos fluidos no miscibles, uno de ellos, un polimero fotocurable que estará dipuesto dentro de un anillo que sirve de soporte donde se analizará su comportamiento en la formación de superficies dentro de una mezcla de agua-glycol$\color{green}{\text{[3]}}$ ,  como se muestra en la $figura 1$. Este análisis se realizará para obtener la curvas minimas que se forman en la superficie de interfase entre los fluidos haciendo uso del calculo variacional se maximizará la función de energía libre del sistema con el fin de resolver el problema isoperimétrico. 

$$Figura\;  1.$$

Para encontrar la energía mínima se requerirá de encontrar los potenciales termodinámicos que intervienen en la formación de la interfase, para esto se realiza un balance de energia al anillo, donde la energia que entra al sistema (anillo con fluido), menos la energía que sale, será igual al cambio de energía del sistema. Este análisis se realizará como un estado estacionario y el balance quedará de la siguiente manera:

$$\begin{equation}
\tag{1.0} E_{in}-E_{out}= \Delta{E_{sist}} \text{.}
\end{equation}$$


La energía que entra al sistema es igual al trabajo $w$ y asumiremos que no existe transferencia de calor $q=0$. El trabajo del sistema estará conformado por tres potenciales, por la tensión superficial del fluído encerrado en el anillo, el potencial gravitatorio y por último el volumen restringido o encerrado dentro del anillo. La suma de las dos primeras energías deberá ser mínima, con la condición de que el volumen $v$ del fuido dentro del anillo permanezca constante.


$$E_{in}= \Delta{E_{sist}} = d{U} = dw \text{.}  \; $$


$$\begin{equation}
\tag{2.0} dU= -2P(dv)+\gamma dA = dw\text{.}
\end{equation}$$


Al aplicar la energía libre de Gibbs $\color{green}{\text{[4]}}$ se tiene que:

$$\begin{equation}
\tag{3.0} G= H  = (U+PV)\text{.}
\end{equation}$$

Donde H representa la entalpía del sistema, se puede obtener la siguiente ecuación diferenciando:

$$\begin{equation}
\tag{3.1} dG=dU+Pdv+v(dP) \text{.}
\end{equation}$$

Al reemplazar la ecuación 2 en 3.1 obtenemos:


$$\begin{equation}
\tag{3.2} dG = \gamma dA  -P(dv)+ vdP \text{.}
\end{equation}$$

La función de Gibbs se puede interpretar como la energía superficial $E_s$, la energia gravitatoria $E_g$ y el volumen restringido $v$.

$$\begin{equation}
\tag{3.3} dG = E_{s} -E_{g}+ \lambda (v) \text{.}
\end{equation}$$  

$$\begin{equation}
\tag{3.4} dG = \gamma dA  -P(dv)+ \lambda(v) \text{.}
\end{equation}$$ 

## Análisis de los terminos de la función de Gibbs

La función de minima energía dependera de la tención superficial, del potencial gravitatorio y del volumen encerrado en el anillo. A continuación dispondremos de los terminos de la función de mínima energía en coordenadas cilindricas como se muestra en la $figura 2$ para el análisis del polimero dentro del anillo de sujeción.

$$figura 2$$.

### Análisis del primer término

El primer término de la ecuación de Gibbs hace referencia a la tensión superfial que actua en un área, este $dA$ se describe de la siguiente manera ecuación 4 $\color{green}{\text{[5]}}$:

$$\begin{equation}
\tag{4.0} dA =\sqrt{1+h_{x}^2+h_{y}²} \; ds \text{.}
\end{equation}$$ 

Donde $h_x$ es la función derivada en función de $x$,  $\frac{\partial h}{\partial x}$ y de $h_y$ como $\frac{\partial h}{\partial y}$, realizaremos el cambio de coordenadas cilindricas para radio $r$ y del angulo $\theta$. Al realizar el análisis observamos que existe simetria rotacional, por lo tanto podemos despreciar el termino en el que el ángulo $\theta$ varia con $x$ y $y$, cambiando las coordenadas queda :  

$$\frac{\partial h}{\partial x} = \frac{\partial h}{\partial r}.\frac{\partial r}{\partial x} + \frac{\partial h}{\partial \theta}.\frac{\partial \theta}{\partial x}\text{,}  \;  \;$$

$$\frac{\partial h}{\partial y} = \frac{\partial h}{\partial r}.\frac{\partial r}{\partial y} + \frac{\partial h}{\partial \theta}.\frac{\partial \theta}{\partial y}\text{,}  \;  \;$$

En coordenadas cilindricas tenemos que: $$\frac{\partial h}{\partial x} = h_r\cos{\theta}+\frac{h_{\theta}\sin{\theta}}{r} \text{,}  \;  \;$$

$$\frac{\partial h}{\partial y} = h_r\sin{\theta}+\frac{h_{\theta}\cos{\theta}}{r}\text{,}  \;  \;$$

Al reemplazar en el $dA$ de la ecuación 4, obtenemos: $$dA=r \sqrt{1+h_{r}^2}\; dr\; d\theta\; \text{,}  \;  \;$$

al integral el primer termino de la función de energía libre de Gibbs :

$$ \gamma dA = \gamma \int_{0}^{R_0} \int_{0}^{2\pi} \sqrt{1+h_{r}^2} \;dr\; d\theta \; \text{,}  \;  \;$$

$$ \gamma dA = \gamma 2\pi\int_{0}^{R_0} \sqrt{1+h_{r}^2}\; dr \text{,} $$

Aplicando el mismo método para la superficie de abajo disponemos de un $h^t$ para la función de la superficie superior y un $ h^b $ para la superficie inferior donde el anillo:
$$\begin{equation}
\tag{4.1} \gamma dA = \gamma 2\pi\int_{0}^{R_0} \left(\sqrt{1+(h_{r}^t)^2} + \sqrt{1+(h_{r}^t)^2}\right) \;dr \text{.}
\end{equation}$$ 

### Análisis del segundo término de la función de Gibbs

El segundo termino de la ecuación de Gibbs en cordenadas polares está dado como:

$$\begin{equation}
\tag{5.0} Pdv = \int_{0}^{R_0} \int_{0}^{2\pi} \Delta \rho g h² r \;dr\; d\theta \; \text{.}
\end{equation}$$  

Al aplicarse tanto para la superficie superior e inferior obtenemos:

$$\begin{equation}
\tag{5.1} Pdv = 2\pi \int_{0}^{R_0}  \Delta \rho g \left((h^t)²- (h^b)²\right)r \;dr \text{.}
\end{equation}$$  

### Análisis del ultimo termino de la función de Gibss.

El último termino de la función de Gibbs en coordenadas polares  para las dos superficies:

$$\begin{equation}
\tag{6.0} v \lambda = \int_{0}^{R_0} \int_{0}^{2\pi}  \lambda_{1}(h^t-h^b) r \;dr\; d\theta\; + \int_{0}^{R_0} \int_{0}^{2\pi}  \lambda_{2}(h^b) r \;dr\; d\theta \text{.}
\end{equation}$$   

$$\begin{equation}
\tag{6.1} v\lambda = 2\pi \left(\int_{0}^{R_0} \lambda_{1} (h^t-h^b) r\; dr +  \int_{0}^{R_0}  \lambda_{2}(h^b) r \;dr \right)\text{.}
\end{equation}$$  

La función de mínima energía (Ec. 3.2) se rescribe de la siguiernte manera: 


$$ G = 2 \pi \gamma \int_{0}^{R_0}  \left(\sqrt{1+(h_{r}^t)^2} + \sqrt{1+(h_{r}^b)^2}- \frac{\Delta \rho g}{\gamma} \left((h^t)²- (h^b)²\right) +\lambda_{1}\frac{(h^t-h^b)}{\gamma}+\lambda_{2}\frac{(h^b)}{\gamma} \right)r\;dr \text{.}$$

$$ G= \int f(r,h,\dot{h})dr \text{.} $$

# Minimización de la energía libre

Miraremos como una función puede minimizarce para entender la superficie minima en nuestro análisis. Dada la función de Gibbs el funcional de pontenciales termodimadinámicos bajo la integral estará dada por la función $f(r,h, h_r,h_{\theta})$  de la siguiente manera:  

$$\begin{equation}
\tag{7} J=\int_{r_0}^{r_1} f\left(r,h,h_{r}, h_{\theta}; r, \theta\right)drd\theta \text{.}
\end{equation}$$

Utilizaremos el metodo de calculo variacional  $\color{green}{\text{[6]}}$ para encontrar la función que nos minimize definiendo $h$ como: 

$$h(r, \theta, ɛ) = h(r,\theta, 0) + ɛH(r,\theta),$$ dónde $H(r,\theta)$ es cero cuando se encuentra en los extremos $r_1$ y $r_2$, para que resulte un extremal ha de cumplirse que $\frac{∂J}{∂ɛ} = 0$.

Que al aplicar esta condición a la ecuación 7 se tiene que:

$$ \frac{\partial J}{\partial ɛ} = \int_{r_0}^{r_1}\left(\frac{\partial}{\partial \varepsilon} f(r,h,h_{r}, h_{\theta}; r, \theta)\right)dr = \int_{r_0}^{r_1}\left(\frac{\partial f}{\partial h}H(r, θ) + \frac{\partial f}{\partial h}\frac{\partial H}{\partial r} + \frac{\partial f}{\partial h_{\theta}}\frac{\partial H}{\partial \theta}\right)dr
\text{.}$$

Nos centramos en el segundo termino de la integral, al evaluar por partes nos que da: $$\int_{r_0}^{r_1}\left(\frac{\partial f}{\partial h}\frac{\partial H}{\partial r}\right)dr = \frac{\partial f}{\partial h}H(r, \theta) \Big |_{r_0}^{r_1} - \int_{r_0}^{r_1}\frac{\partial }{\partial r} \left(\frac{\partial f }{\partial h_r} \right)H(r, \theta)dr \text{.} $$

Como $H(r, \theta)$ se anula en los extremos la primera parte de la integral es 0, siguiendo el mismo razonamiento para el tercer termino de la integral original, se llega a: $$\frac{\partial J}{\partial ϵ}=\int_{r_0}^{r_1}\left(\frac{\partial f}{\partial h} -\frac{\partial }{\partial r} \frac{\partial f }{\partial h_r} - \frac{\partial}{\partial \theta}\frac{\partial f}{\partial h_{\theta}}\right)H(r, \theta)dr=0 \text{.} $$

Por lo tanto: 
$$\begin{equation}
\frac{\partial f}{\partial h} -\frac{\partial }{\partial r} \frac{\partial f }{\partial h_r} - \frac{\partial}{\partial \theta}\frac{\partial f}{\partial h_{\theta}}=0 \text{.}
\end{equation}$$


Al existir simetria de rotación podemos omitir el ultimo término dando como resultado la ecuación de Euler-lagrange:


$$\begin{equation}
\tag{8.0} \frac{\partial f}{\partial h} -\frac{\partial }{\partial r} \frac{\partial f }{\partial h_r} =0 \text{.}
\end{equation}$$

La solución por medio de la ecuación de Euler-lagrange se lograra realizandose dos veses para las dos superficies dando como resultrado un sistema de ecuaciones diferenciales acopladas en terminos de $h^t$ y $h^b$ donde f es: 

$$\begin{equation}
f= \left( \sqrt{1+(h_{r}^t)^2} + \sqrt{1+(h_{r}^b)^2}- \frac{\Delta \rho g}{\gamma} \left((h^t)²- (h^b)²\right) +\lambda_{1}\frac{(h^t-h^b)}{\gamma}+\lambda_{2}\frac{(h^b)}{\gamma}\right) \, r\;dr \text{.}
\end{equation}$$

```python
from sympy import *
from sympy.solvers.pde import pdsolve

import numpy as np
```

```python
def f(h_t, h_b, h_rt,h_rb, h_0t, h_0b, r,
      delta_P = Symbol("\\Delta P"), g = Symbol("g"), gamma = Symbol("\\gamma"),
      lambda1 = Symbol("\\lambda_1"), lambda2 = Symbol("\\lambda_2")):
  
  return (sqrt(1 + h_rt**2 + h_0t**2/r**2) + sqrt(1 + h_rb**2 + h_0t**2/r**2) - delta_P*g/(2*gamma)*(h_t**2 - h_b**2) + lambda1*(h_t - h_b)/gamma + lambda2*h_b/gamma)*r

r, theta = symbols("r \\theta")

#Caso h(r, theta).--------
h_t = Function("h^t")(r, theta); h_rt = diff(h_t, r);  h_0t = diff(h_t, theta)
h_b = Function("h^b")(r,theta); h_rb = diff(h_t, theta);  h_0b = diff(h_b, theta)
funcional = f(h_t, h_b, h_rt,h_rb, h_0t, h_0b, r)

#Caso independiente de theta h(r).
h_t = Function("h^t")(r); h_rt = diff(h_t, r);
h_b = Function("h^b")(r); h_rb = diff(h_b, r)
funcional = f(h_t, h_b, h_rt,h_rb, 0, 0, r) #Funcional.

#Ecuaciones diferenciales apartir de Euler-Lagrange.
PDE1 = Eq(diff(funcional, h_t) - simplify(diff(diff(funcional, h_rt), r)) - simplify(diff(diff(funcional, h_0t), theta), 0)) #PDE para h_t
PDE2 = Eq(diff(funcional, h_b) - simplify(diff(diff(funcional, h_rb), r)) - simplify(diff(diff(funcional, h_0b), theta), 0)) #PDE para h_b
```
